```python
diseases,all_symptomes = [],[]
for x in tqdm(range(len(json_dict))):
    df = pd.DataFrame(json_dict[str(x)]['disease'].items(),columns=["disease","percentage"])
    disease = df.sort_values("percentage",ascending= False).iloc[0]["disease"]
    symptomes = ",".join(json_dict[str(x)]['symptoms'][0])
    
    #symptomes
#     tqdm.write(f"{disease} : {symptomes}\n")
    diseases.append(disease)
    all_symptomes.append(symptomes)
```

```python
 zipped = list(zip(all_symptomes,diseases))
    
#temp df
temp_df = pd.DataFrame(zipped,columns=["symptomes",'disease'])
temp_df
```

```python
def remove_esc_chars(string):
    review = re.sub('''[^a-zA-Z0-9!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~°]''', ' ',string)
    review = review.split()
    review = ' '.join(review)
    return review.strip()
```

```python
temp_df.drop_duplicates(inplace=True)
temp_df.sort_values("symptomes",inplace=True)
temp_df["disease"] = temp_df["disease"].apply(lambda x : remove_esc_chars(x))
temp_df["symptomes"] = temp_df["symptomes"].apply(lambda x : remove_esc_chars(x))
```

```python
temp_df[temp_df.duplicated()].sort_values("symptomes").tail(20)
```

```python
#this will cut the dataframe into parts 

last = 0
for x in range(0,len(temp_df),1500):
    if x == 0:
        continue
    temp_df[last:x].to_csv(f'disease and symptomes{x}.csv',index=False)
    last = x
```
