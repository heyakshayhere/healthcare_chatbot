```python
import os 
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import pandas as pd,requests,bs4,re,time,io,pytesseract,easyocr,random,textstat,urllib.request
from pdfminer.high_level import extract_text
from PIL import Image
from pathlib import Path
from pdf2image import convert_from_path
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from goose3 import Goose
from datetime import datetime
from bs4 import BeautifulSoup
from selenium import webdriver
from tqdm.auto import tqdm

reader = easyocr.Reader(['en'])

import warnings
warnings.filterwarnings("ignore")

%autosave 1
```

```python
SITE_NAME = "NHSinform"

DOMAIN = "https://www.nhsinform.scot"

SITE_LINK="https://www.nhsinform.scot/illnesses-and-conditions/a-to-z"
```

```python
#driver for operation
from webdriver_manager.chrome import ChromeDriverManager
option = webdriver.ChromeOptions()
# option.add_argument('headless')
driver = webdriver.Chrome(ChromeDriverManager().install(),options=option)
```

```python
def parse_webpage_bs(search_url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:90.0) Gecko/20100101 Firefox/90.0"}
    try:
        site_request = requests.get(search_url, headers=headers, timeout=10)
    except requests.exceptions.RequestException as e:
        print(e)
        site_request = None
    if site_request != None and site_request.status_code==200:
        site_soup = bs4.BeautifulSoup(site_request.content, "lxml")
    else:
        site_soup = None
    return site_soup
```

```python
soup = parse_webpage_bs(SITE_LINK)
container = soup.find('div', {'class': 'col small-12 panel-content'})
```

```python
elements = container.findAll('ul', {'class': 'nhsuk-list nhsuk-list--border'})
```

```python
diseases,links = [],[]
for element in elements:
    try:
        elem = element.findAll('li')
        for e in elem:
            disease = e.text.strip()
            if disease != "Back to top":
                diseases.append(disease)
                print(disease)
                href = DOMAIN + e.find('a')['href']
                links.append(href)
                print(href)
    except:
        pass
```

```python
#validating data shape
len(diseases),len(links),len(diseases) == len(links)
```

```python
texts = []
for link in tqdm(links[:]):
    driver.get(link)
    
    #cookies
    try:
        driver.find_element_by_xpath('''//*[@id="ccc-notify-accept"]''').click()
    except:
        pass

    text = ""
    try:
        for x in range(100):
            ps = driver.find_element(By.XPATH,f'''//*[@id="maincontent"]/div[2]/div/div[2]/div/div[1]/div[2]''').text.strip()
            text += ps
            driver.find_element_by_xpath('''//a[@class="js-guide__incremental--previous nav-incremental-link page-navigation__next"]''').click()
    except:
        pass
    texts.append(text)
```

```python
#zippig all the data togather 
zipped = list(zip(diseases,texts,links))
```

```python
data_list = []

for disease,text,link in zipped:
    article = (disease.strip(),text.strip(),link.strip())
    data_list.append(article)
    print(disease)
```

```python
temp_df = pd.DataFrame(data_list,columns=['disease','text','url'])
temp_df
```

```python
def see_data(iloc_no=random.randint(0,len(temp_df))-1):
    print(temp_df.iloc[iloc_no]['disease'],temp_df.iloc[iloc_no]['url'])
    print(f"\n{temp_df.iloc[iloc_no]['text']}")

see_data()
```

```python
#to csv
temp_df.to_csv(f'{SITE_NAME} disease_data.csv',index = False)
```
